```python
#Imports
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 150
import numpy as np
import math
from scipy import stats
sns.set()
```

```python
pd.options.display.width = 1000
```

```python
parasweep = pd.read_csv("fox_rabbit3D_multipara.csv")
```

```python
parasweep
```

```python
# The distribution of running times
sns.histplot(data=parasweep,x='runtime')
plt.show()
```

```python
# The distribution of runtimes equal to or greater than 20 seconds
para_over_20 = parasweep[parasweep["runtime"]>=20]
sns.histplot(data=para_over_20,x='runtime')
plt.show()
```

```python
# The distribution of runtimes less than 20 seconds
para_under_20 = parasweep[parasweep["runtime"]<20]
sns.histplot(data=para_under_20,x='runtime')
plt.show()
```

```python
#  The distribution of runtimes for parameter sweeps dataset
sns.scatterplot(y=para_under_20.index, x = para_under_20["runtime"],s = 1.5)
sns.scatterplot(y=para_over_20.index, x = para_over_20["runtime"],s = 1.5)
plt.ylabel("Simulation index")
```
