```python
#Imports
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.rcParams['figure.dpi'] = 150
import numpy as np
import math
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
sns.set_style("whitegrid")
```

```python
# Parameter sweeps dataset
parasweep = pd.read_csv("fox_rabbit3D_multipara.csv")
```

```python
parasweep
```

```python
# Parameter sweeps, displayed with three parameters
fig = plt.figure(figsize = (12,8))
fig.tight_layout()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=parasweep["r_birth"], ys=parasweep["f_birth"], zs=parasweep["f_death"],marker="*", s=0.1)
plt.tick_params(labelsize=8)
ax.set_xlabel("r_birth",fontsize=12)
ax.set_ylabel("f_birth",labelpad=7,fontsize=12)
ax.set_zlabel("f_death",labelpad=-1.7,fontsize=12)
```

```python
# Parameter sweeps, displayed with two parameters
sns.set()
sns.scatterplot(x=parasweep["r_birth"], y=parasweep["f_birth"],  s=10)
```
